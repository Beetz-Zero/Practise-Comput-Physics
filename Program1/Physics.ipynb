```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('Out_1.txt')

energy = data[:, 0]  
x = data[:, 1]       
y = data[:, 2]       

sc = plt.scatter(x, y, c=energy)

plt.colorbar(sc, label='Уровень энергии')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.xticks(np.arange(-5, 6, 1))
plt.yticks(np.arange(-5, 6, 1))
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
x1=x.reshape(-1,int(np.sqrt(x.size)))
y1=y.reshape(-1,int(np.sqrt(x.size)))
energy1=energy.reshape(-1,int(np.sqrt(x.size)))
```

```python
plt.pcolormesh(x1,y1,energy1,cmap='viridis', shading='auto')
plt.colorbar()
```
